```python
import os
import shutil
import numpy as np

# Directory where your images and labels are stored
data_dir = '/Users/nisargp/PycharmProjects/CV/Assignment /VinBigDatasetSmall/Train_Watershed'

# Directories for the split datasets
train_images_dir = '/Users/nisargp/PycharmProjects/CV/Assignment /Train Images'
val_images_dir = '/Users/nisargp/PycharmProjects/CV/Assignment /Val Images'
test_images_dir = '/Users/nisargp/PycharmProjects/CV/Assignment /Test Images'
train_labels_dir = '/Users/nisargp/PycharmProjects/CV/Assignment /Train Images Labels'
val_labels_dir = '/Users/nisargp/PycharmProjects/CV/Assignment /Val Images Labels'
test_labels_dir = '/Users/nisargp/PycharmProjects/CV/Assignment /Test Images Labels'

# Create these directories if they don't exist
for directory in [train_images_dir, val_images_dir, test_images_dir, train_labels_dir, val_labels_dir, test_labels_dir]:
    os.makedirs(directory, exist_ok=True)

# Parameters for splitting the dataset
train_ratio = 0.7
val_ratio = 0.2
# Test ratio is implied to be 0.1 as they should sum up to 1

# Get all image filenames
image_filenames = [f for f in os.listdir(data_dir) if f.endswith('.jpg')]  # Change this if your images are not in jpg format
np.random.shuffle(image_filenames)  # Shuffle the list to ensure random distribution

# Compute split indices
total_images = len(image_filenames)
train_end = int(train_ratio * total_images)
val_end = train_end + int(val_ratio * total_images)

# Function to copy files
def copy_files(filenames, src_dir, dst_dir_images, dst_dir_labels):
    for filename in filenames:
        # Construct the full file paths
        src_image_path = os.path.join(src_dir, filename)
        dst_image_path = os.path.join(dst_dir_images, filename)
        # Assuming label files match image files exactly but with .txt extension
        src_label_path = os.path.join(src_dir, os.path.splitext(filename)[0] + '.txt')
        dst_label_path = os.path.join(dst_dir_labels, os.path.splitext(filename)[0] + '.txt')
        
        # Copy files
        shutil.copy(src_image_path, dst_image_path)
        shutil.copy(src_label_path, dst_label_path)

# Split and copy files
copy_files(image_filenames[:train_end], data_dir, train_images_dir, train_labels_dir)
copy_files(image_filenames[train_end:val_end], data_dir, val_images_dir, val_labels_dir)
copy_files(image_filenames[val_end:], data_dir, test_images_dir, test_labels_dir)
```

```python
import os
import numpy as np
import tensorflow as tf
from skimage.io import imread
from skimage.transform import resize
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, concatenate, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, concatenate, Conv2DTranspose, Dropout, BatchNormalization
from tensorflow.keras.models import Model

# Define the data loading function
def load_data(images_path, labels_path, img_height=256, img_width=256):
    images = []
    masks = []
    image_filenames = [f for f in os.listdir(images_path) if f.endswith('.jpg')]
    
    for filename in image_filenames:
        img_path = os.path.join(images_path, filename)
        img = imread(img_path, as_gray=True)
        img = resize(img, (img_height, img_width), mode='constant', preserve_range=True)
        
        label_filename = os.path.splitext(filename)[0] + '.txt'  # Matching label file
        label_path = os.path.join(labels_path, label_filename)
        
        if os.path.exists(label_path):  # Check if corresponding label file exists
            try:
                mask = np.loadtxt(label_path)  # Adjust depending on the format of your labels
                if mask.ndim == 1:  # Ensure mask is not flat when it should be 2D
                    mask = np.expand_dims(mask, axis=0)
                mask = resize(mask, (img_height, img_width), mode='constant', preserve_range=True)
                mask = np.expand_dims(mask, axis=-1)  # Add channel dimension
                images.append(img)
                masks.append(mask)
            except Exception as e:
                print(f"Could not process {filename} due to: {e}")
        else:
            print(f"Label file for {filename} not found.")

    # Ensure images and masks are numpy arrays and normalized
    images = np.array(images) / 255.0
    masks = np.array(masks) / 255.0  # Assuming masks are in a format that can be normalized like this
    return images, masks


# Define the U-Net model
def unet(input_size=(256, 256, 1)):
    inputs = Input(input_size)
    # Downsample
    c1 = Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    p1 = MaxPooling2D((2, 2))(c1)
    c2 = Conv2D(128, (3, 3), activation='relu', padding='same')(p1)
    p2 = MaxPooling2D((2, 2))(c2)
    
    # Bottleneck
    c3 = Conv2D(256, (3, 3), activation='relu', padding='same')(p2)
    
    # Upsample
    u4 = UpSampling2D((2, 2))(c3)
    c4 = Conv2D(128, (3, 3), activation='relu', padding='same')(u4)
    u5 = UpSampling2D((2, 2))(c4)
    c5 = Conv2D(64, (3, 3), activation='relu', padding='same')(u5)
    
    # Output
    outputs = Conv2D(1, (1, 1), activation='sigmoid')(c5)
    model = Model(inputs=[inputs], outputs=[outputs])
    return model

# Paths to your data
train_images_path = '/Users/nisargp/PycharmProjects/CV/Assignment /Train Images'
train_labels_path = '/Users/nisargp/PycharmProjects/CV/Assignment /Train Images Labels'
val_images_path = '/Users/nisargp/PycharmProjects/CV/Assignment /Val Images'
val_labels_path = '/Users/nisargp/PycharmProjects/CV/Assignment /Val Images Labels'

# Load datasets
train_images, train_masks = load_data(train_images_path, train_labels_path)
val_images, val_masks = load_data(val_images_path, val_labels_path)

# Initialize the U-Net model
model = unet()
model.compile(optimizer=Adam(learning_rate=1e-4), loss='binary_crossentropy', metrics=['accuracy'])

# Setup callbacks
callbacks = [
    EarlyStopping(patience=10, verbose=1),
    ModelCheckpoint('unet_best_model.h5', verbose=1, save_best_only=True)
]

# Train the model
results = model.fit(
    train_images, train_masks,
    validation_data=(val_images, val_masks),
    batch_size=8,
    epochs=10,
    callbacks=callbacks
)
```

```python
import matplotlib.pyplot as plt

def plot_training_history(history):
    # Plot the training and validation loss
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss Over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Plot the training and validation accuracy
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy Over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Assuming 'results' is your history object from training
plot_training_history(results)
```

```python
def plot_model_predictions(model, images, true_masks, num=5):
    plt.figure(figsize=(15, 10))

    # Randomly select 'num' images
    indices = np.random.choice(range(len(images)), num, replace=False)

    for i, index in enumerate(indices):
        img = np.expand_dims(images[index], axis=0)  # Model expects batch dimension
        pred_mask = model.predict(img)[0]  # Get the first (and only) prediction
        pred_mask = np.squeeze(pred_mask)  # Remove batch dimension

        # Plot original image
        plt.subplot(num, 3, i * 3 + 1)
        plt.imshow(images[index].squeeze(), cmap='gray')
        plt.title('Original Image')
        plt.axis('off')

        # Plot true mask
        plt.subplot(num, 3, i * 3 + 2)
        plt.imshow(true_masks[index].squeeze(), cmap='gray')
        plt.title('True Mask')
        plt.axis('off')

        # Plot predicted mask
        plt.subplot(num, 3, i * 3 + 3)
        plt.imshow(pred_mask, cmap='gray')
        plt.title('Predicted Mask')
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Assuming you have 'val_images' and 'val_masks' as your validation datasets
plot_model_predictions(model, val_images, val_masks, num=5)
```

```python
import numpy as np
from tensorflow.keras.models import load_model
model = load_model('unet_best_model.h5')

def iou(y_true, y_pred):
    intersection = np.logical_and(y_true, y_pred)
    union = np.logical_or(y_true, y_pred)
    iou_score = np.sum(intersection) / np.sum(union)
    return iou_score

def dice_coefficient(y_true, y_pred):
    intersection = np.logical_and(y_true, y_pred)
    return 2. * np.sum(intersection) / (np.sum(y_true) + np.sum(y_pred))

predicted_masks = model.predict(val_images)
# Threshold predictions
predicted_masks[predicted_masks > 0.5] = 1
predicted_masks[predicted_masks <= 0.5] = 0

# Calculate IoU and Dice for each image
iou_scores = [iou(true, pred) for true, pred in zip(val_masks, predicted_masks)]
dice_scores = [dice_coefficient(true, pred) for true, pred in zip(val_masks, predicted_masks)]

# Average IoU and Dice for the validation set
avg_iou = np.mean(iou_scores)
avg_dice = np.mean(dice_scores)

print(f'Average IoU: {avg_iou}, Average Dice Coefficient: {avg_dice}')
```
